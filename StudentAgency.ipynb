**Scores of Student Agency of Lesson Plans Generated Across AI Tools**

```python
import matplotlib.pyplot as plt

# Data
key_constructs = [
    "Authentic Problems", "Goal-Directed", "Make Choices", "Take Actions",
    "Problem Solving", "Self-Efficacy", "Interaction", "Collaboration",
    "Opportunities to Share Ideas", "Play Different Roles", "Share Authority"
]

scores_magic_school = [24, 0, 2, 0, 23, 22, 29, 25, 30, 2, 0]
scores_school_ai = [17, 0, 2, 0, 11, 22, 30, 26, 28, 1, 0]
scores_vanilla_ai = [22, 0, 1, 0, 20, 19, 28, 14, 23, 0, 0]
scores_custom_ai = [29, 25, 29, 26, 29, 28, 30, 26, 30, 25, 25]

# Dimensions
dimensions = [
    "Dispositional", "Dispositional", "Motivational", "Motivational",
    "Motivational", "Motivational", "Positional", "Positional",
    "Positional", "Positional", "Positional"
]

# Line styles and colors
colors = ["red", "blue", "purple", "green"]
line_styles = ["--", "-.", ":", "-"]

# Create the plot
plt.figure(figsize=(14, 8))

# Plot for each platform
platforms = {
    "Magic School": scores_magic_school,
    "School AI": scores_school_ai,
    "Vanilla AI": scores_vanilla_ai,
    "Custom AI": scores_custom_ai,
}

for i, (platform, scores) in enumerate(platforms.items()):
    plt.plot(key_constructs, scores, line_styles[i], color=colors[i], label=platform)
    for j, score in enumerate(scores):
        plt.annotate(score, (key_constructs[j], score), textcoords="offset points", xytext=(0, 5), ha='center')

# Highlighting dimensions with background colors
for idx, dimension in enumerate(["Dispositional", "Motivational", "Positional"]):
    dimension_indices = [i for i, d in enumerate(dimensions) if d == dimension]
    start, end = min(dimension_indices), max(dimension_indices)
    plt.axvspan(start - 0.5, end + 0.5, alpha=0.1 * (idx + 1), color=colors[idx])

# Labels and title
plt.xlabel("Key Constructs", fontweight="bold")
plt.ylabel("Scores", fontweight="bold")
plt.title("Scores of Student Agency of Lesson Plans Generated Across AI Tools", fontweight="bold")
plt.xticks(rotation=45, ha='right')
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
