**Student Agency Score Comparison Across Platforms**

```python
import matplotlib.pyplot as plt

# Topics and scores for each platform
topics = [
    "The Number System", "Expressions and Equations-1", "Expressions and Equations-2",
    "Earth and Space Science", "Life Science", "Physical Science", "Engineering and Technology Engineering Design"
]

magic_school_scores = [6, 5, 4, 13, 11, 10, 6]
school_ai_scores = [8, 4, 7, 5, 6, 5, 6]
our_generator_scores = [4, 4, 4, 2, 4, 5, 4]

# Plotting
plt.figure(figsize=(12, 6))

# Plot Magic School scores
plt.plot(topics, magic_school_scores, '--', color='red', label='Magic School')
for i, score in enumerate(magic_school_scores):
    plt.annotate(score, (topics[i], score), textcoords="offset points", xytext=(0, 5), ha='center')

# Plot School AI scores
plt.plot(topics, school_ai_scores, '-.', color='blue', label='School AI')
for i, score in enumerate(school_ai_scores):
    plt.annotate(score, (topics[i], score), textcoords="offset points", xytext=(0, 5), ha='center')

# Plot Our Generator scores
plt.plot(topics, our_generator_scores, '-', color='green', label='Our Generator')
for i, score in enumerate(our_generator_scores):
    plt.annotate(score, (topics[i], score), textcoords="offset points", xytext=(0, 5), ha='center')

# Labels and title
plt.xlabel("Topics")
plt.ylabel("Student Agency Score")
plt.title("Student Agency Score Comparison Across Platforms", fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
